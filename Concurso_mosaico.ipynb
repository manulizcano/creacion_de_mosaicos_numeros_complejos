<h1>Vamos a contruir una lista de números complejos llamada "lista_base" cuya imagen en el plano complejo va a ser una "R". </h1>





```python
lista_base = [] # aquí vamos a guardar la R
for n in range(100):                                     #
  t = (n/100)                                            #
  lista_base.append(t*1j)                                #
  lista_base.append(t*(0.5j )+(1-t))                     # Construimos la R
  lista_base.append(1j+(1-t))                            #
  lista_base.append(0.5j+(1-t))                          #
  lista_base.append((0.5j+1)*t+(1j+1)*(1-t))             #

lista = np.array(lista_base) #volvemos nuestra lista de complejos en un vector para poder extraer la parte real e imaginaria
lista = (lista -(0.5 +0.5j))*0.5 + (0.5 +0.5j) # escalamos nuestra R
lista_base = lista.tolist()

parte_real = lista.real # extraemos la parte real
parte_imaginaria = lista.imag # extraemos la parte imaginaria

plt.scatter(parte_real, parte_imaginaria)  # de aquí para abajo es para graficar
plt.ylabel('Parte imaginaria')
plt.xlabel('Parte real')
plt.show()
```

<h1>Observemos que podemos crear un mosaico muy fácilmente sumando a cada punto de nuestra "R" complejos x+yi con x,y enteros.</h1>

```python
lista_mosaico = []
n = 10
m = 7

r = np.array(lista_base)
for x in range(n):
  for y in range(m):
    r_traladada = r +(x+y*1j)
    lista_mosaico += r_traladada.tolist()

mosaico = np.array(lista_mosaico)
parte_real = mosaico.real
parte_imaginaria = mosaico.imag

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.scatter(parte_real, parte_imaginaria)
plt.ylabel('Parte imaginaria')
plt.xlabel('Parte real')
plt.show()
```
